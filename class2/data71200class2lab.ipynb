```python
# https://www.kaggle.com/datasets/prevek18/ames-housing-dataset
import pandas as pd
from sklearn.datasets import fetch_openml
housing = fetch_openml(name="house_prices", as_frame=True)
```

```python
housing_df=pd.DataFrame(housing.data)
```

```python
print(housing_df.head())
```

```python
housing_df['SalePrice'] = housing.target
```

```python
%matplotlib inline
import matplotlib.pyplot as plt

housing_df.hist(bins=50, figsize=(20,15))
plt.show()
```

```python
from pandas.plotting import scatter_matrix

attributes = ["SalePrice", "LotArea", "1stFlrSF","YearBuilt"]
scatter_matrix(housing_df[attributes], figsize=(12, 8))
```

```python
corr_matrix = housing_df[attributes].corr()
corr_matrix["SalePrice"].sort_values(ascending=False)
```
